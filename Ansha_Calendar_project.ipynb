### Step 04 : Create a list 'cal_mth' with Month dates as strings and leading values as blanks
* **For any Calendar month your code should be able to run and generate the correct output. Use the below guidelines :**
* For dates < 10, add leading Zeroes to make the String length = 2
* The blank values in the in the beginning of the list 'cal_mth' (as shown in the cell output below) have length 2 each. Essentially these are Double spaces
* The count of these blank values in the List will depend on DOW (Day of Week) of First Day of the month. E.g.
 * _If First Day of the month falls on a Monday, then no blank values should appear_
 * _If First Day of the month falls on a Tuesday, then 1 blank value should appear_
 * _If First Day of the month falls on a Wednesday, then 2 blank values should appear_
 * _..._
 * _If First Day of the month falls on a Sunday, then 6 blank values should appear_

```python
  #---------Pick different month names here to test the code's output-------
def generate_calendar_month(year, month):
    import calendar

    # Get the calendar for the given month and year
    cal = calendar.monthcalendar(year, month)

    # Calculate the number of blank values to add to the beginning of the list
    first_day_of_month = calendar.weekday(year, month, 1)
    num_blanks = first_day_of_month

    # Initialize the calendar month list with blank values
    cal_mth = ['  '] * num_blanks

    # Iterate over the calendar days and add them to the list
    for week in cal:
        for day in week:
            if day == 0:
                # If the day is zero, it means it's a blank space in the calendar
                cal_mth.append('  ')
            else:
                # Add a leading zero if the day is less than 10
                cal_mth.append(str(day).zfill(2))

    return cal_mth
#-----------Write you code here--------------
#-----------Write you code here--------------
#-----------Write you code here--------------


#----------Last line in this cell should display the created List--------
```
